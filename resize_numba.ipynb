```python
@jit(nopython=True)
def resize(image , rate):
    rO, cO, dO = image.shape
    new_image = np.zeros((int(rO * rate[0]), int(cO * rate[1]), dO))
    new_image = new_image.astype('uint8')
    rN, cN, dN = new_image.shape
    
    for idxR in range(rN):
        for idxC in range(cN):
            for idxD in range(dN):
                new_image[idxR, idxC, idxD] = image[int(idxR/rate[0]), int(idxC/rate[1]), idxD]
    return new_image

def resize2(image , rate):
    rO, cO, dO = image.shape
    new_image = np.zeros((int(rO * rate[0]), int(cO * rate[1]), dO))
    new_image = new_image.astype('uint8')
    rN, cN, dN = new_image.shape
    
    for idxR in tqdm(range(rN)):
        for idxC in range(cN):
            for idxD in range(dN):
                new_image[idxR, idxC, idxD] = image[int(idxR/rate[0]), int(idxC/rate[1]), idxD]
    return new_image
```
